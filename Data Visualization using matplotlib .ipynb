```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('mtcars.csv')
```

```python
df.head()
```

```python
plt.plot('model','hp',data=df)
plt.xlabel('Models of the cars')
plt.ylabel(' Horse-Power of Cars')
plt.title('Model Names vs HorsePower')
plt.show()
```

```python
plt.bar('carb','gear',data=df)
plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward gears')
plt.title('carbs vs gear')
plt.show()
```

```python
plt.hist('wt',bins=30,data=df)
plt.xlabel('Weight of cars')
plt.ylabel('count')
plt.title('histogram for the weight values')
plt.show()
```

```python
models=df['model']
plt.pie('cyl',labels=models,shadow=True,data=df)
plt.title('cylinders')
plt.show()
```

```python
am=df['am']
carb=df['carb']
fig=plt.figure(figsize=(10,6))
ax1=plt.subplot()
ax1.stackplot(am,carb)
plt.xlabel('transmission')
plt.ylabel('number of caburetors')
plt.title('transmission vs carburetors')
```
