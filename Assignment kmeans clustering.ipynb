## Assignment 1

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('kmeans.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['Age'],df['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans
```

```python
x=df.iloc[:,2:5]
```

```python
x.head()
```

```python
cluster=KMeans()
```

```python
cluster.fit(x)
```

```python
identifiedclusters=cluster.predict(x)
```

```python
identifiedclusters
```

```python
data_with_clusters=df.copy()
data_with_clusters['clusters']=identifiedclusters
plt.scatter(data_with_clusters['Age'],data_with_clusters['Spending Score (1-100)'],c=data_with_clusters['clusters'],cmap='rainbow')
plt.show()
```

## assignment 2

```python
wcss=[]
for i in range(1,10):
    kmeans=KMeans(i)
    kmeans.fit(x)
    wcss_iter=kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters=range(1,10)
plt.plot(number_clusters,wcss)
plt.title('elbow method')
plt.show()
```

```python
cluster=KMeans(6)
cluster.fit(x)
```

```python
identifiedclusters=cluster.predict(x)
```

```python
data_with_clusters=df.copy()
data_with_clusters['clusters']=identifiedclusters
plt.scatter(data_with_clusters['Age'],data_with_clusters['Spending Score (1-100)'],c=data_with_clusters['clusters'],cmap='rainbow')
plt.show()
```
